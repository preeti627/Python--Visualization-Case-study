# Q1. Compare Sales by region for 2016 with 2015 using bar chart

```python
dt= sales_data.groupby("Region")["Sales2015","Sales2016"].sum()
dt
```

```python
dt.plot(kind= 'bar',figsize=(7,5))
```

```python
print("we can conclude that sales in 2016 was always higher than sales in 2015 and region with most sales in both year is EAST")
```

# Q2. What are the contributing factors to the sales for each region in 2016. Visualize it using a Pie Chart.

```python
df=sales_data.groupby(by= 'Region').Sales2016.sum().reset_index()
```

```python
plt.figure(figsize=(6,6))
plt.pie('Sales2016' , labels='Region',data= df, autopct="%.2f%%")
plt.show()
```

```python
print("The most contributing factor to the sales in 2016 is EAST region with 42.70% sales")
```

# Q6. Create a new column Qtr using numpy.where() or any suitable utility in the imported dataset. The Quarters are based on months and defined as -
* Jan - Mar : Q1
* Apr - Jun : Q2
* Jul - Sep : Q3
* Oct - Dec : Q4

```python
sales_data.loc[(sales_data.Month== 'Jan')|(sales_data.Month== 'Feb') | (sales_data.Month== 'Mar' ),'Qtr'] = 'Q1'
sales_data.loc[(sales_data.Month== 'Apr')|(sales_data.Month== 'May' )| (sales_data.Month== 'Jun'),'Qtr'] = 'Q2'
sales_data.loc[(sales_data.Month== 'Jul')|(sales_data.Month== 'Aug' )|( sales_data.Month== 'Sep'),'Qtr'] = 'Q3'
sales_data.loc[(sales_data.Month== 'Oct')|(sales_data.Month== 'Nov' )|( sales_data.Month== 'Dec' ),'Qtr'] = 'Q4'
```

# Q7. Compare Qtr wise sales in 2015 and 2016 in a bar plot

```python
d=sales_data.groupby('Qtr')["Sales2015","Sales2016"].sum()
```

```python
d.plot(kind='bar')
```

```python
print("In above chart, sales of Qtr in 2016 is more than sales of quarter in 2015.")
```

# Q8. Determine the composition of Qtr wise sales in and 2016 with regards to all the Tiers in a pie chart.

```python
q=sales_data[(sales_data.Tier=='High')|(sales_data.Tier=='Low')|(sales_data.Tier=='Med')].groupby(['Qtr','Tier']).Sales2016.sum().reset_index()
q= pd.pivot_table(data= q,index='Tier', values='Sales2016', columns='Qtr')
```

```python
q.plot(kind= 'pie',x= 'Sales2016',subplots= True, figsize= (20,20), autopct="%.2f%%")
```

```python
print("Above chart clearly shows that in every quarter, tier (HIGH) is more than the other two",
      "combined , i.e. always above 50 percent and in Q3 &Q4 it is more than 60%.")
```
